```python
import psycopg2
import os
from pprint import pprint
conn = psycopg2.connect(host="localhost",database='postgres',user=os.environ["POSTGRES_USER"],password=os.environ["POSTGRES_PASSWORD"])

with conn:
    with conn.cursor() as cursor:
        #取出最新日期各站點資料
        sql = '''
            SELECT 日期,站點資訊.站點名稱,行政區,站點地址,lat,lng,總車輛,可借,可還,可借,活動
            FROM youbike
            JOIN 站點資訊 ON youbike.編號 = 站點資訊.站點編號
            WHERE (日期,編號) IN (
                SELECT MAX(日期),編號
                FROM youbike
                GROUP BY 編號
            );
        '''
        cursor.execute(sql)
        allData = cursor.fetchall()
        pprint(allData)

conn.close()
```
